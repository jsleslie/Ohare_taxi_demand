```python
import pandas as pd
import numpy as np
from sklearn.dummy import DummyRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

### Manual Modelling

```python
ORD_train_lag = pd.read_csv('../data/train_data.csv', index_col='Unnamed: 0')
ORD_train_lag = ORD_train_lag.dropna()
ORD_train_lag.head(1)

y_train = ORD_train_lag['rides']
X_train = ORD_train_lag.drop(columns=['rides'])
```

```python
sk_dr = DummyRegressor(strategy='mean')
sk_rf = RandomForestRegressor()
sk_lr = LogisticRegression(solver='liblinear', penalty='l1')
sk_gbm = GradientBoostingRegressor()
```

Best autoML score: 38.3853

```python
np.sqrt(-1*np.mean(cross_val_score(sk_dr, X_train, y_train, scoring='neg_mean_squared_error')))
```

```python
np.sqrt(-1*np.mean(cross_val_score(sk_rf, X_train, y_train, scoring='neg_mean_squared_error')))
```

```python
np.sqrt(-1*np.mean(cross_val_score(sk_lr, X_train, y_train, scoring='neg_mean_squared_error')))
```

```python
np.sqrt(-1*np.mean(cross_val_score(sk_gbm, X_train, y_train, scoring='neg_mean_squared_error')))
```
