```python
#STARTING WITH 84 ZIPFILES, MAKE ORD_OTP.CSV

import pandas as pd
import numpy as np
import zipfile as zp
```

```python
entries = []

for i in range(84):
    if i == 0:
        with zp.ZipFile('../data/otp/1051953426_T_ONTIME_REPORTING.zip') as myzip:
            myzip.extract('1051953426_T_ONTIME_REPORTING.csv',path='../data/otp')
            df = pd.read_csv('../data/otp/1051953426_T_ONTIME_REPORTING.csv')
            entries.append(df.query('DEST == "ORD"'))
            
    else:
        with zp.ZipFile('../data/otp/1051953426_T_ONTIME_REPORTING '+'('+str(i)+').zip') as myzip:
            myzip.extract('1051953426_T_ONTIME_REPORTING.csv', path ='../data/otp')
            df = pd.read_csv('../data/otp/1051953426_T_ONTIME_REPORTING.csv')
            entries.append(df.query('DEST == "ORD"'))
```

```python


len(entries)
```

```python
combined_ord = pd.concat(entries)
```

```python
combined_ord = combined_ord.drop(columns=['ORIGIN_AIRPORT_ID','ORIGIN_AIRPORT_SEQ_ID',
                                          'ORIGIN_CITY_MARKET_ID','DEST_CITY_MARKET_ID',
                                          'Unnamed: 15'])
```

```python
combined_ord['ARR_HOUR'] = combined_ord['ARR_TIME']//100
```

```python
combined_ord.to_csv('../data/ORD_OTP.csv')
```
